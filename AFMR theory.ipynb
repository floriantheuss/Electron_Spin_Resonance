# easy axis

```python
def w1 (B, g, Ba):
    return g * ( B + Ba ) 

def w2 (B, g, Ba):
    return g * ( - B + Ba ) 

def w3 (B, g, Ba):
    return g * np.sqrt( B**2 - Ba**2 ) 

def w4 (B, g, Ba):
    return g * np.sqrt( B**2 + Ba**2 )

def w5 (B, g, Ba, Be):
    return g * Ba * np.sqrt( 1 - B**2/Be**2 )
```

```python
g = 15
Ba = 6
Be = 25

B = np.linspace(0,16,100000)
B1 = [i for i in B[B<=Ba]]
B2 = [i for i in B[B>=Ba]]

B1nan = [i for i in np.zeros(len(B)-len(B1))*np.nan]
B2nan = [i for i in np.zeros(len(B)-len(B2))*np.nan]

B1 = np.array(B1 + B1nan)
B2 = np.array(B2nan + B2)



plt.plot(B, w1(B1, g, Ba))
plt.plot(B, w2(B1, g, Ba))
plt.plot(B, w3(B2, g, Ba))
plt.plot(B, w4(B, g, Ba))
plt.plot(B, w5(B, g, Ba, Be))
```

```python
data = [B, w1(B1, g, Ba), w2(B1, g, Ba), w3(B2, g, Ba), w4(B, g, Ba), w5(B, g, Ba, Be)]
names = ['', '', '', '','', '']
units = ['', '', '', '','', '']
comments = ['', '', '', '','', '']

filename = 'C:\\Users\\F25_1.307_b\\Box Sync\\Klingeler_Masterarbeit\\LaTex\\chapter2_theory\\data_for_figures\\AFMR_easy_axis.txt'

save(data, names, units, comments, filename)
```

# easy plane

```python
def w1 (B, g):
    return g * B

def w2 (B, g, Ba, Be):
    return g * np.sqrt( 2*Ba*Be - (Ba/(2*Be)) * B**2 )

def w3 (B, g, Ba, Be):
    return g * np.sqrt( 2*Ba*Be + (1 - Ba/(2*Be)) * B**2 )
```

```python
g = 15
Ba = 1.9
Be = 10

B = np.linspace(0,16,100000)
plt.plot(B, w1(B, g))
plt.plot(B, w2(B, g, Ba, Be))
plt.plot(B, w3(B, g, Ba, Be))
```

```python
data = [B, w1(B, g), w2(B, g, Ba, Be), w3(B, g, Ba, Be)]
names = ['', '', '', '']
units = ['', '', '', '']
comments = ['', '', '', '']

filename = 'C:\\Users\\F25_1.307_b\\Box Sync\\Klingeler_Masterarbeit\\LaTex\\chapter2_theory\\data_for_figures\\AFMR_easy_plane.txt'

save(data, names, units, comments, filename)
```
