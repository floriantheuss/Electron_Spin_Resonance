```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate
import math
import os.path
%matplotlib inline

# import the SQUID data
filepath = 'C:\\Users\\F25_1.307_b\\Nextcloud\\Klingeler_Masterarbeit\\howardevansite\\NaCuFe2(VO4)3 powder'
filename = 'NaCuFe2(VO4)3_fixed powder_M(T)_zfc_5T'

# mass of the sample in mg
mass = 2.1*3.5

# magnetic field in Oe
b=[]
# temperature in K
t=[]
# long moment in emu
m=[]
# standard deviation of long moment
deltam=[]

f = open(filepath+'\\'+filename+'.dat', 'r')
f.readline()
for line in f:
    line=line.strip()
    line=line.split()
    b.append(float(line[0]))
    t.append(float(line[1]))
    m.append(float(line[2]))
    deltam.append(float(line[3]))   

b = np.array(b)
t = np.array(t)
m = np.array(m)
deltam = np.array(deltam)

# since two measurements are taken at the same temperature we will average over those
B = (b[::2]+b[1::2])/2
T = (t[::2]+t[1::2])/2
M = (m[::2]+m[1::2])/2/mass
deltaM = (deltam[::2]+deltam[1::2])/2

# take a discrete derivative of M(T)
derivativeM = []
for i in np.arange(len(B)):
    if i ==0:
        derivativeM.append((M[i+1]-M[i])/(T[i+1]-T[i]))
    elif 0 < i and i < len(B)-1:
        derivativeM.append(((M[i]-M[i-1])/(T[i]-T[i-1])+(M[i+1]-M[i])/(T[i+1]-T[i]))/2)
    else:
        derivativeM.append((M[i]-M[i-1])/(T[i]-T[i-1]))

derivativeM = np.array(derivativeM)



# write the data into a new file or overwrite an existing file
newfilename=filepath+'\\'+filename+'_processed.dat'

if os.path.isfile(newfilename) == True:
    x='w'
else:
    x='x'

with open(newfilename, x) as g:
    g.write('Temperature'+'\t'+'Field'+'\t'+'Magnetization'+'\t'+'Long Scan Std Dev'+'\t'+'dM/dT'+'\n')
    g.write('K'+'\t'+'Oe'+'\t'+'erg/(Gs mg)'+'\t'+'emu'+'\t'+'erg/Gs mg K)'+'\n')
    for i in np.arange(len(B)):
        g.write(str(T[i])+'\t'+str(B[i])+'\t'+str(M[i])+'\t'+str(deltaM[i])+'\t'+str(derivativeM[i])+'\n')
```
